```python
# Import dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store CSV file in dataframe

csv_file = "hate_data.csv"
hate_df = pd.read_csv(csv_file)
hate_df.head()
```

```python
# Rename key columns

hate_df.rename(columns = {'STATE_NAME':'State', 'DATA_YEAR':'Year', 'VICTIM_COUNT':'Victim_Count'}, inplace = True)
hate_df
```

### Create new data with select columns

```python
# Create new dataframe only with key columns and print

new_hate_df = hate_df[['Year', 'State', 'Victim_Count']].copy()
new_hate_df
```

```python
# Group by State and Year and count Victims, then print

hate_group_df = new_hate_df.groupby(["State", 'Year']).count()
hate_group_df
```

```python
# Store CSV file in dataframe

csv_file = "min_wage.csv"
df = pd.read_csv(csv_file,encoding="cp1252")
df
```

```python
# Using Loc function to match the years with the previus dataframe. This study will cover only years from 1991 to 2018
# Print

wage_df =  df.loc[(df["Year"] >= 1991) & (df["Year"] <= 2018)]
wage_df
```

```python
# Rename key columns

wage_df.rename(columns = {'State.Minimum.Wage':'State_Minimum_Wage', 'Federal.Minimum.Wage':'Federal_Minimum_Wage'}, inplace = True)
```

```python
# Create new dataframe only with key columns and print

new_min_wage_df = wage_df[['State', 'Year','State_Minimum_Wage', 'Federal_Minimum_Wage']].copy()
new_min_wage_df.head()
```

```python
# Reset the index

new_min_wage_df = new_min_wage_df.reset_index(drop=True)
new_min_wage_df
```

```python
# Merge the two dataframes based in the State & Year including all the data, then print

merge_df = pd.merge(new_min_wage_df, hate_group_df, on=('State','Year'), how='outer')
merge_df
```

```python
# Convert NaN in zero value 

merge_df['Victim_Count'] = merge_df['Victim_Count'].fillna(0)
```

```python
# Drop every row called Federal in the State column, then print

merge_df = merge_df[merge_df['State'] != 'Federal']
merge_df
```

```python
# Store the merged file into a CSV file

merge_df.to_csv("merge_df.csv")
```
